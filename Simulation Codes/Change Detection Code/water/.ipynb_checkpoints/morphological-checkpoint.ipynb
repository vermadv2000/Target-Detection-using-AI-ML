```python
import sys
import numpy as np
import cv2 as cv2
from skimage import io
from PIL import Image
import imutils
import matplotlib.pylab as plt
```

```python
image = io.imread("nn_change.png")
plt.imshow(image, cmap='gray')
plt.show()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
kernel1 = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
kernel2 = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
kernel3 = cv2.getStructuringElement(cv2.MORPH_RECT, (3,1))
kernel4 = cv2.getStructuringElement(cv2.MORPH_RECT, (1,3))
```

```python
image1=image
opening = cv2.morphologyEx(image1, cv2.MORPH_OPEN, kernel)
plt.imshow(opening, cmap='gray')
plt.show()
```

```python
image1=opening
eroded = cv2.erode(image1, kernel, iterations = 3)
plt.imshow(eroded, cmap='gray')
plt.show()
```

```python
image1=image
dilated = cv2.dilate(opening, kernel4, iterations = 2)
plt.imshow(dilated, cmap='gray')
plt.show()
```

```python
image1=image
dilated2 = cv2.dilate(dilated, kernel3, iterations = 1)
plt.imshow(dilated2, cmap='gray')
plt.show()
```

```python
f, axarr = plt.subplots(1,4)
axarr[0].imshow(image, cmap='gray')

axarr[1].imshow(opening, cmap='gray')
axarr[2].imshow(dilated, cmap='gray')
axarr[3].imshow(dilated2, cmap='gray')
plt.show()
```
