```python
import rasterio
from rasterio.plot import show
import os.path
import numpy as np
import pandas as pd
from PIL import Image
```

```python
img1 = rasterio.open("18_14band_svm.png")
img2 = rasterio.open("22_14band_svm.png")

show(img1)
show(img2)
```

```python
z1 = img1.read()
print(z1.shape)
z2 = img2.read()
print(z2.shape)
```

```python
a = np.array(z1)
pred1 = a.reshape(104,76)
rslt = np.unique(pred1)
print(rslt)
a = np.array(z2)
pred2 = a.reshape(104,76)
rslt = np.unique(pred2)
print(rslt)
temp1 = pred2 - pred1

for i in range(104):
    for j in range(76):
        if(temp1[i][j]==1):
            temp1[i][j]=0

rslt = np.unique(temp1)

print(rslt)
img3 = Image.fromarray(temp1)
img3.save('svm_change.png' )
show(img3)
```
