```python
import rasterio
from rasterio.plot import show
import os.path
import numpy as np
import pandas as pd
from PIL import Image
```

```python
img1 = rasterio.open("rf18_14band.png")
img2 = rasterio.open("rf22_14band.png")

show(img1)
show(img2)
```

```python
z1 = img1.read()
print(z1.shape)
z2 = img2.read()
print(z2.shape)
```

```python
a = np.array(z1)
pred1 = a.reshape(104,76)

a = np.array(z2)
pred2 = a.reshape(104,76)

temp1 = pred2 - pred1

img3 = Image.fromarray(temp1)
img3.save('rf_change.png')
show(img3)
```
