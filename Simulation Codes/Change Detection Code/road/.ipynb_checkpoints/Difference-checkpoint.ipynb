```python
import rasterio
from rasterio.plot import show
import os.path
import numpy as np
import pandas as pd
from PIL import Image
```

```python
img1 = rasterio.open("nn18_try2.png")
img2 = rasterio.open("nn22_15band.png")

show(img1, cmap='gray')
show(img2, cmap='gray')
```

```python
z1 = img1.read()
print(z1.shape)
z2 = img2.read()
print(z2.shape)
```

```python
a = np.array(z1)
pred1 = a.reshape(173,118)

a = np.array(z2)
pred2 = a.reshape(173,118)

temp1 = pred2 - pred1

img3 = Image.fromarray(temp1)
img3.save('diff2.tif')
show(img3)
```
